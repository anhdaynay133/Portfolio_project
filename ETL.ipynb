```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt 
```

```python
path ='D:/Kỳ 20211/Đồ án 2/Datasource/'
data = pd.read_csv(path + 'vehicles.csv')
data
```

```python
data.info()
```

```python
for col in data.columns:
    missing_data = data[col].isna().sum()
    missing_percen = missing_data / len(data) * 100
    print(f"Column {col}: has {missing_percen}% missing data")
```

### Xóa các cột có nhiều giá trị NULL

```python
data = data.drop(['county','size'], axis = 1)
```

```python
data.columns
```

```python
data['Quantity'] = np.random.randint(1, 10, data.shape[0])
data.head()
```

### Thêm cột Sales 


```python

data['Sales'] = data['Quantity'] * data['price']
data.head()
```

### Xóa các hàng mà các cột có giá trị NULL

```python
data.dropna(subset = ["VIN",'drive','type','paint_color','condition','cylinders'], inplace=True)
```

```python
data.info()
```

### Thêm cột Date 


```python
data["Order_date"] = np.random.choice(pd.date_range('2016-01-01', '2021-01-01'), len(data))
```

### Di chuyển cột


```python
moving_columns = data.pop('Order_date')
data.insert(5,'Order_date',moving_columns)
```

```python
data.head(3)
```

### Tách và thêm cột ngày tháng năm từ Order_date

```python
data['Day'] = data['Order_date'].dt.day
data['Month'] = data['Order_date'].dt.month
data['Year'] = data['Order_date'].dt.year
```

```python
data.head(3)
```
